```python
# Ecuacion cinemática de y en funcion de x, se halló despéjando el tiempo
# de la ecuacion cinematica de x cuando x=distancia de tiro 

vo=float(input("Ingrese la Velocidad inicial: ")) 

θ=float(input("Ingrese el ángulo de tiro: "))
yo=float(input("Ingrese su altura: "))
d=float(input("Ingrese la distancia del tiro: "))

if d<0 or vo<0 or θ<0 or yo<0 :
  print("La distancia, el ángulo y la velocidad inicial no pueden ser negativos")

else:
  θ=np.radians(θ)
  g=9.8
  h=3.05
  r=0.15
  y1=yo+np.tan(θ)*(d-r)-(1/2)*g*((d-r)/(vo*np.cos(θ)))**2      # Altura considerando una distancia d-r
  y2=yo+np.tan(θ)*d-(1/2)*g*(d/(vo*np.cos(θ)))**2              # Altura considerando una distancia d
  y3=yo+np.tan(θ)*(d+r)-(1/2)*g*((d+r)/(vo*np.cos(θ)))**2      # Altura considerando una distancia d+r
  t_v=(vo*np.sin(θ)+np.sqrt(((vo*np.sin(θ))**2)+2*g*yo))/g     # Tiempo de vuelo

  xmax=xmax=vo*np.cos(θ)*t_v    # Distancia máxima en x

  # Condicion para no hacer cesta por debajo
  if y1<y3 and y1<h:
    print("No puede hacer canasta por debajo")
  else:
    if y1>h and y3 <h:     # Condición para la cesta considerando un margen de radio r
      print ("!Cestaaaaa OMG!")

 # Gráfica     
x=np.arange(0,xmax,xmax/100) # Para plotear toda la curva
y=yo+np.tan(θ)*(x)-(1/2)*g*((x)/(vo*np.cos(θ)))**2 # y en función de x

plot(d,h,"o",label="Centro canasta")
plot(d-r,h,"o",label="Inicio canasta")
plot(d+r,h,"o",label="Final canasta")
plot(x,y)
xlabel("Eje X")
ylabel("Eje Y")
legend(loc='lower left')
savefig("Trayectoria.jpg")
```
